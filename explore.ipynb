ANALISIS DE VARIABLES

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Buscar variables categoricas
categoricas = df1.select_dtypes(include=['object']).columns.tolist()
categoricas
```

```python
categoricas_seleccionadas = ['GENDER', 'RURAL', 'TYPE OF ADMISSION-EMERGENCY/OPD', 'OUTCOME', 'CHEST INFECTION']
```

```python
# Crear el gráfico con el número adecuado de filas y columnas (9 filas, 2 columnas)
num_rows = (len(categoricas_seleccionadas) + 1) // 2  # +1 para asegurarnos de redondear hacia arriba
fig, axis = plt.subplots(num_rows, 2, figsize=(15, 2 * num_rows))

# Recorrer las columnas categóricas y generar los countplots
for i, cat in enumerate(categoricas_seleccionadas):
    row, col = i // 2, i % 2  # Determinar fila y columna
    sns.countplot(ax=axis[row, col], data=df1, x=cat)  # Graficar el countplot
    plt.setp(axis[row, col].get_xticklabels(), rotation=45, ha='right')  # Rotar las etiquetas del eje X

plt.tight_layout()  # Ajustar el espaciado entre los subgráficos
plt.show()
```

```python
numericas = df1.select_dtypes(include=['int']).columns.tolist()
numericas
```

```python
# Suponiendo que 'numericas' es una lista de columnas numéricas
fig, axis = plt.subplots(38, 2, figsize=(20, 40))  # Aumentando el tamaño de la figura

for i, num in enumerate(numericas):
    if i % 2 == 0:
        sns.histplot(ax=axis[i, 0], data=df1, x=num).set(xlabel=None)
        sns.boxplot(ax=axis[i + 1, 0], data=df1, x=num)
    else:
        sns.histplot(ax=axis[i - 1, 1], data=df1, x=num).set(xlabel=None)
        sns.boxplot(ax=axis[i, 1], data=df1, x=num)

# Ajustar el espaciado entre los gráficos
plt.subplots_adjust(hspace=0.4, wspace=0.2)  # Espaciado horizontal y vertical

# Ajustar la distribución de las subgráficas
plt.tight_layout()  # Asegura que las etiquetas y títulos no se solapen

# Mostrar la gráfica
plt.show()
```
