```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Buscar variables categoricas
categoricas = df1.select_dtypes(include=['object']).columns.tolist()
categoricas
```

```python
categoricas_seleccionadas = ['GÃ©nero', 'Residencia', 'Tipo de AdmisiÃ³n', 'Motivo de Alta', 'InfecciÃ³n TorÃ¡cica', 'REINGRESO']
```

```python
fig, axis = plt.subplots(3, 2, figsize = (10, 10))

for i, cat in enumerate(categoricas_seleccionadas):
    row, col = i // 2, i % 2
    sns.countplot(ax=axis[row,col], data=df1[categoricas], x=cat)
    plt.setp(axis[row, col].get_xticklabels(), rotation=45, ha='right')
    
plt.tight_layout()
plt.show()
```

```python
# Ajustar el tamaÃ±o de la figura
plt.figure(figsize=(40, 40))  # Cambia los valores para modificar el tamaÃ±o

# Crear el heatmap
sns.heatmap(df_numerico.corr(), annot=True, fmt='.2f', cmap='coolwarm')

# Mostrar el grÃ¡fico
plt.title('Mapa de calor de correlaciones')
plt.show()
```

```python
correlation_matrix = df_numerico.corr()

# Seleccionar las correlaciones que no sean la diagonal principal (1.0)
correlation_pairs = correlation_matrix.unstack()
correlation_pairs = correlation_pairs[correlation_pairs != 1]

# Ordenar por valor absoluto de la correlaciÃ³n
sorted_correlation = correlation_pairs.abs().sort_values(ascending=False)

# Mostrar las correlaciones mayores (puedes ajustar el umbral)
threshold = 0.25  # Define un umbral mÃ­nimo para mostrar correlaciones significativas
high_correlation = sorted_correlation[sorted_correlation > threshold]

print("Variables con mayor correlaciÃ³n:")
print(high_correlation)

# Opcional: Graficar un heatmap filtrado con las correlaciones mayores al umbral
plt.figure(figsize=(10, 10))
sns.heatmap(
    correlation_matrix[(correlation_matrix.abs() > threshold) & (correlation_matrix != 1)],
    annot=True, fmt='.2f', cmap='coolwarm'
)
plt.title('Variables con alta correlaciÃ³n')
plt.show()
```
